## Softmax

To turn the scores $S$ to probabilities.

$$S(y_i)=\dfrac{e^{y_i}}{\sum_{j=1}^{n}e^{y_j}}$$

Scores are also called logits.

```python
# import numpy as np
import numpy as np

def softmax(x):
    """Compute softmax values for each sets of scores in x."""
    # Quiz 1
    return np.exp(x) / np.exp(x).sum(axis=0)

scores = [3.0, 1.0, 0.2]
print(softmax(scores))

# Plot softmax curves
import matplotlib.pyplot as plt
x = np.arange(-2.0, 6.0, 0.1)
scores = np.vstack([x, np.ones_like(x), 0.2 * np.ones_like(x)])

plt.plot(x, softmax(scores).T, linewidth=1)
plt.show()
```

```python
# Quiz 2
scores = np.array([3.0, 1.0, 0.2])
# If multiply by 10
print("softmax(scores*10)")
print(softmax(scores*10))
# Probability gets close to either 0.0 or 1.0
# Increase the size of the input, increase confidence of the output

# Plot softmax curves
import matplotlib.pyplot as plt
x = np.arange(-2.0, 6.0, 0.1)
scores = np.vstack([x, np.ones_like(x), 0.2 * np.ones_like(x)])

plt.plot(x, softmax(scores*10).T, linewidth=1)
plt.show()
```

```python
# Quiz 3
scores = np.array([3.0, 1.0, 0.2])
# If divided by 10
print("softmax(scores/10)")
print(softmax(scores/10))
# Probability gets close to uniform distribution. 
# Since all the scores decrease in magnitude, the resulting softmax probabilities will be closer to each other.
# Decrease the size iof the input, decrease the confidenceof the output

# Plot softmax curves
import matplotlib.pyplot as plt
x = np.arange(-2.0, 6.0, 0.1)
scores = np.vstack([x, np.ones_like(x), 0.2 * np.ones_like(x)])

plt.plot(x, softmax(scores/10).T, linewidth=1)
plt.show()
```

## Cross Entropy

$$D({S,L}) = -\sum_{i=1}^{n} {L_i, \log(S_i)} $$

S is the Classifer matrix.

L is the Label vector (target value).

D is the Distance (cost function). 

Not symmetric, $$D({S,L}) \ne D({L,S})$$

D has a very big gradient when the target value is 1 and the output is almost zero.
